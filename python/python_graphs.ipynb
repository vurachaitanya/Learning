```python
import pandas
import matplotlib.pyplot as plt 
import seaborn as sns 

data=pandas.read_csv(r"C:\Users\vchaita3\Desktop\files\Jupyter Notebook\datasets\Data Sets\Case Study based Datasets\Bank_churn_modelling.csv")
plt.figure(figsize=(12,8))
plt.subplot(211)
sns.countplot(data['HasCrCard'])
plt.subplot(212)
sns.countplot(data['HasCrCard'][data['Exited']==1])
plt.show()

### Having credit card is no impact
```

```python
plt.figure(figsize=(12,5))
sns.factorplot(x='Geography',col='HasCrCard',data=data[data.Exited==1],kind='count')
plt.show()
## Having credicard is not making much diffrence in different countries 
```

```python
#Gender is impacting
#If yes which Gender has specific issues
#If Yes is gender having idfferent impact is different Geographies?
sns.countplot(data['Gender'][data['Exited']==1])
plt.show()
```

```python
plt.figure(figsize=(12,8))
plt.subplot(211)
sns.countplot(data.Gender)
plt.subplot(212)
sns.countplot(data['Gender'][data['Exited']==1])
plt.show()
## Female has some issues as leaving are more
```

```python
plt.figure(figsize=(12,5))
sns.factorplot(x='Geography',col='Gender',data=data[data.Exited==1],kind='count')
plt.show()
#Males leaving Germany is leaving are high 
```

```python
cor=data.corr()
plt.figure(figsize=(12,5))
sns.heatmap(cor,annot=True,cmap='coolwarm')
plt.show()

#Creditscore vs Exited = -0.027 which is not impacting 
###Age is increasing leaving is also increasing +ve corelation then it is impacting who are leaving is more------Y
####Tenure -0.014 will not have impact ---N
###Balance is +ve, So people who are with good balance probability of leaving is more 0.12
#Balance vs no of products --those who are having high balance will take much loans. 
#high Balance ---leaving
#Less Balance --- High products

```
