```python
import numpy
import matplotlib.pyplot as plt
x=numpy.arange(0,30)
y=numpy.sin(x)

plt.figure(figsize=(12,5))
plt.scatter(x,y,label='Sine(x)') ## Only Dots
plt.legend()
plt.title('Graph for Sinx')
plt.xlabel('Value of X')
plt.ylabel('Value of Singx')
plt.plot(x,y) # Connecting the Dots
plt.show()
```
