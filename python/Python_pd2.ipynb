```python
import pandas
data=pandas.read_csv(r"C:\Users\vchaita3\Desktop\files\Jupyter Notebook\datasets\Data Sets\regiment.csv")
print (data.head())
print('\n')
print(data['preTestScore'].mean())
print('\n')

print(data['regiment'].unique().size)
print('\n')

print(data['regiment'].unique())
print('\n')


data.groupby(by ='regiment')['preTestScore'].mean()
print('\n')


data.groupby(by ='regiment')['preTestScore','postTestScore'].mean()
print('\n')

data.groupby(by = 'company')['preTestScore','postTestScore'].mean()
print('\n')

data.groupby(by = ['company','regiment'])['preTestScore'].mean()
```

```python
import pandas
data=pandas.read_csv(r"C:\Users\vchaita3\Desktop\files\Jupyter Notebook\datasets\Data Sets\Case Study based Datasets\Bank_churn_modelling.csv")

#Data which dose not impact the customer leaving the account are 
#CustomerID, Surename, RowNumber, 
## Drop which are not require
data.drop(['RowNumber','CustomerId','Surname'],axis=1,inplace=True)
### if row should be removed  AXIS=0
#### if Column should be removed AXIS=1
#data=data.iloc[:,3:] #########other option
####Age is impacting or not & Check if any missing values
print(data.isnull().sum())

### impact of age on people leaving or not leaving
print('\n')
print(data['Age'][data['Exited']==1].mean()) ## Who are still customers
print(data['Age'][data['Exited']==0].mean()) #who left

print('\n')
print(data['Age'].describe())
```

```python
import pandas
import matplotlib.pyplot as plt
import seaborn as sns
data=pandas.read_csv(r"C:\Users\vchaita3\Desktop\files\Jupyter Notebook\datasets\Data Sets\Case Study based Datasets\Bank_churn_modelling.csv")

plt.figure(figsize=(12,5))
sns.distplot(data['CreditScore'][data['Exited']==1])
sns.distplot(data['CreditScore'][data['Exited']==0])
plt.legend(['1','0'])
plt.show()
##CReditscore is not impacting the score of people who are leaving 
```

```python
import numpy
import matplotlib.pyplot as plt
import seaborn as sns

x=numpy.arange(0,30)
y=numpy.sin(x)
z=numpy.cos(x)

#plt.plot(x,y,z)
plt.figure(figsize=(12,4))
plt.plot(x,y,'r',label='sin(x)')
plt.plot(x,z,'g',label='cos(x)')
plt.legend()
plt.show()
```
